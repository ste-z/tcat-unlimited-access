```python


# Option A: Quick start with a Developer Token
# Generate this in the Box Developer Console under your app's 'Configuration' tab
client = DevelopmentClient() 

# The public URL of the Box file or folder
shared_url = 'https://app.box.com'

# Get the shared item (can be a File or Folder object)
shared_item = client.get_shared_item(shared_url)

# If it's a file, download its content
if shared_item.type == 'file':
    with open(shared_item.name, 'wb') as open_file:
        shared_item.download_to(open_file)
    print(f"Downloaded: {shared_item.name}")
```
